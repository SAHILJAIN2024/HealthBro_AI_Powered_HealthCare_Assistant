```python
import pandas as pd 
dataset1 = pd.read_csv("Advanced patient dataset.csv")
dataset1.head(10)
```

```python
dataset1.drop('Patient_ID', axis=1 , inplace = True)
```

```python
dataset1.head(10)
```

```python
dataset2 = pd.read_csv("unstructured_patient_dataset.csv")
dataset2.head(10)
```

```python
dataset2["Symptoms"] = dataset2["Symptoms"].str.lower().str.strip()
dataset2.dropna(inplace=True)
```

```python
dataset2.head(10)
```

```python
dataset = pd.concat([dataset1, dataset2], ignore_index=True)
dataset
```

```python
from sklearn.utils import shuffle
dataset= shuffle(dataset, random_state=42)
dataset
```

```python
from sklearn.feature_extraction.text import TfidfVectorizer

tfidf = TfidfVectorizer(max_features=3000, stop_words='english', ngram_range=(1,2))
X = tfidf.fit_transform(dataset["Symptoms"].values.astype("U"))
X
```

```python
import numpy as np
from sklearn.preprocessing import LabelEncoder, OneHotEncoder
from sklearn.compose import ColumnTransformer

le_severity = LabelEncoder()
y_severity = le_severity.fit_transform(dataset["Severity"])

ohe = OneHotEncoder(sparse_output=False)
y_doctor = ohe.fit_transform(dataset[["Suggested Doctor"]])

Y = np.hstack((y_severity.reshape(-1, 1), y_doctor))
# Y = np.vstack((y_severity, y_doctor)).T
Y
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42)
```

```python
from sklearn.multioutput import MultiOutputClassifier
from xgboost import XGBClassifier
model = MultiOutputClassifier(XGBClassifier(use_label_encoder=False, eval_metric='mlogloss'))
model.fit(X_train, Y_train)
```

```python
from sklearn.metrics import accuracy_score, classification_report

y_pred = model.predict(X_test)
print("Severity Accuracy:", accuracy_score(Y_test[:, 0], y_pred[:, 0]))
print("Doctor Accuracy:", accuracy_score(Y_test[:, 1], y_pred[:, 1]))

print("\n🔹 Severity Classification Report:")
print(classification_report(Y_test[:, 0], y_pred[:, 0]))

print("\n🔹 Doctor Classification Report:")
print(classification_report(Y_test[:, 1], y_pred[:, 1]))
```

```python
# Use argmax to find the predicted class index
y_true_doctor = np.argmax(Y_test[:, 1:], axis=1)
y_pred_doctor = np.argmax(y_pred[:, 1:], axis=1)

# Labels (doctor names)
doctor_labels = ohe.categories_[0]

# Confusion Matrix
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

cm = confusion_matrix(y_true_doctor, y_pred_doctor)

plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues',
            xticklabels=doctor_labels,
            yticklabels=doctor_labels)
plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Doctor Prediction')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
from sklearn.metrics import confusion_matrix
import matplotlib.pyplot as plt
import seaborn as sns

# Get true and predicted severity labels (already label-encoded)
y_true_severity = Y_test[:, 0]
y_pred_severity = y_pred[:, 0]

# Define label names if needed (optional)
severity_labels = le_severity.classes_

# Plot confusion matrix
cm_severity = confusion_matrix(y_true_severity, y_pred_severity)

plt.figure(figsize=(6, 5))
sns.heatmap(cm_severity, annot=True, fmt='d', cmap='Greens',
            xticklabels=severity_labels,
            yticklabels=severity_labels)

plt.xlabel('Predicted')
plt.ylabel('Actual')
plt.title('Confusion Matrix - Severity Prediction')
plt.tight_layout()
plt.show()
```

```python
import joblib

# Save model components
joblib.dump(model, "structured_model.pkl")
joblib.dump(tfidf, "structured_vectorizer.pkl")
joblib.dump(le_severity, "structured_severity_encoder.pkl")
joblib.dump(ohe, "structured_doctor_encoder.pkl")
```

```python
def predict_symptom(symptom_text):
    # 1. Lowercase input
    symptom_text = symptom_text.lower()

    # 2. TF-IDF transform
    X = tfidf.transform([symptom_text])

    # 3. Predict with model
    preds = model.predict(X)

    # 4. Split predictions
    severity_encoded = preds[0][0]
    doctor_encoded = (preds[0][1:] > 0.5).astype(int)
    # Make sure doctor_encoded is reshaped properly
    doctor_encoded = np.array(doctor_encoded).reshape(1, -1)

    # 5. Decode predictions
    severity_label = le_severity.inverse_transform([severity_encoded])[0]
    try:
        doctor_label = ohe.inverse_transform(doctor_encoded)[0][0]
    except:
        doctor_label = "Unknown Doctor (Prediction unclear)"

    # 6. Return result
    return {
        "severity": severity_label,
        "suggested_doctor": doctor_label
    }
```

```python
result = predict_symptom("sore throat, weight loss, headache, dizziness")
print(result)
```
